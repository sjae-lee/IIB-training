```python
import matplotlib.pyplot as plt
import numpy as np
import scipy as sp
```

### Maximum likelihood estimation

```python
fig, ax = plt.subplots(1,3,figsize=(15,5))
ax[0].plot(np.linspace(16,30,100),sp.stats.norm.pdf(np.linspace(16,30,100),loc=22,scale=1),c='k')
ax[0].text(16,0.7,'Likelihood: %0.3f'%sp.stats.norm.pdf(23,loc=22,scale=1))
ax[1].plot(np.linspace(16,30,100),sp.stats.norm.pdf(np.linspace(16,30,100),loc=26,scale=1),c='k')
ax[1].text(16,0.7,'Likelihood: %0.3f'%sp.stats.norm.pdf(23,loc=26,scale=1))
ax[2].plot(np.linspace(16,30,100),sp.stats.norm.pdf(np.linspace(16,30,100),loc=23,scale=1),c='k')
ax[2].text(16,0.7,'Likelihood: %0.3f'%sp.stats.norm.pdf(23,loc=23,scale=1))

for i in range(3):
    ax[i].vlines(23,0,1,ls='--',colors='grey', linewidth=0.5)
    ax[i].set_xlabel('$T$ [$\\degree$C]',fontsize=12)
    ax[i].set_xticks(range(16,31,2))
    ax[i].set_xticklabels(range(16,31,2))
    ax[i].set_xlim(15.5, 30.5)
    ax[i].set_ylim(0, 0.75)
```

### Bayesian Sequential Update

Suppose that your prior is 
$$
T_{true}\sim\mathcal{N}(20,3)
$$
You have one measurement 
$$T_{meas}=21$$
The measurement uncertainty follows 
$$T_{meas}\sim\mathcal{N}(T_{true},1)$$
Show the posterior distribution. 

```python
x = 21

fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.vlines(x,0,1,ls='--',colors='grey', linewidth=0.5)

mean_prior=20
sd_prior=3
sd_sensor=1
ax.plot(np.linspace(15,30,100),sp.stats.norm.pdf(np.linspace(15,30,100),loc=mean_prior,scale=sd_prior),c='k')
ax.text(16,0.7,"Prior: $\\mathcal{N}(%d,%d)$"%(mean_prior,sd_prior))

######################### posterior #########################
mean_posterior=1/(1/sd_prior**2 + 1/sd_sensor**2) * (mean_prior/sd_prior**2 + x/sd_sensor**2)
sd_posterior=(((1/sd_prior**2 + 1/sd_sensor**2)**-1) )**0.5
ax.plot(np.linspace(15,31,200),sp.stats.norm.pdf(x=np.linspace(15,31,200),loc=mean_posterior, scale=sd_posterior),'C0')
ax.text(16,0.65,"Posterior: $\\mathcal{N}(%.2f,%.2f)$"%(mean_posterior,sd_posterior))
##################################################

ax.set_xlabel('$T$ [$\\degree$C]',fontsize=12)
ax.set_xticks(range(16,31,2))
ax.set_xticklabels(range(16,31,2))
ax.set_xlim(15.5, 30.5)
ax.set_ylim(0, 0.75)
```

Now, we have one more measurement:
$$
T_{meas}=25
$$
Show the updated posterior

```python
x=25

fig, ax = plt.subplots(1,1,figsize=(5,5))
ax.vlines(x,0,1,ls='--',colors='grey', linewidth=0.5)

mean_prior=mean_posterior
sd_prior=sd_posterior
sd_sensor=1
ax.plot(np.linspace(15,30,100),sp.stats.norm.pdf(np.linspace(15,30,100),loc=mean_prior,scale=sd_prior),c='k')
ax.text(16,0.7,"Prior: $\\mathcal{N}(%d,%d)$"%(mean_prior,sd_prior))

######################### posterior #########################
mean_posterior=1/(1/sd_prior**2 + 1/sd_sensor**2) * (mean_prior/sd_prior**2 + x/sd_sensor**2)
sd_posterior=(((1/sd_prior**2 + 1/sd_sensor**2)**-1) )**0.5
ax.plot(np.linspace(15,31,200),sp.stats.norm.pdf(x=np.linspace(15,31,200),loc=mean_posterior, scale=sd_posterior),'C0')
ax.text(16,0.65,"Posterior: $\\mathcal{N}(%.2f,%.2f)$"%(mean_posterior,sd_posterior))
##################################################

ax.set_xlabel('$T$ [$\\degree$C]',fontsize=12)
ax.set_xticks(range(16,31,2))
ax.set_xticklabels(range(16,31,2))
ax.set_xlim(15.5, 30.5)
ax.set_ylim(0, 0.75)
```
