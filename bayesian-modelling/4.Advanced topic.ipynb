```python
try:
    import GPy
except:
    %pip install Gpy
    import GPy

import numpy as np
import matplotlib.pyplot as plt
```

```python
def function(x):
    return np.sin(x-5) + np.cos(x*1.6+4.7)*0.5 + x*0.15 - 0.5
x_plot = np.linspace(0,10,100)
y_plot = function(x_plot)
```

```python
np.random.seed(1234)
plt.subplots()
plt.plot(x_plot,y_plot)
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(0,10)
plt.ylim(-2,2)
```

```python
plt.subplots()
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(0,10)
plt.ylim(-2,2)
```

```python
x = np.array([-10])[:,None]

kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
m = GPy.models.GPRegression(x,function(x),kernel,noise_var=0)

plt.subplots()
plt.fill_between(x_plot,m.predict_quantiles(x_plot[:,None])[0].flatten(),m.predict_quantiles(x_plot[:,None])[1].flatten(),alpha=0.3)
plt.plot(x_plot,m.predict(x_plot[:,None])[0])
plt.plot(x_plot,y_plot,'--',c='C1')
plt.scatter(x,function(x),marker='x',c='r',s=80)
plt.xlabel('x',fontsize=12)
plt.ylabel('y',fontsize=12)
plt.xlim(-0,10)
plt.ylim(-2,2)

x = np.array([0])[:,None]
for i in range(5):
    plt.subplots()
    plt.fill_between(x_plot,m.predict_quantiles(x_plot[:,None])[0].flatten(),m.predict_quantiles(x_plot[:,None])[1].flatten(),alpha=0.3)
    plt.plot(x_plot,m.predict(x_plot[:,None])[0])
    plt.plot(x_plot,y_plot,'--',c='C1')
    plt.scatter(x,function(x),marker='x',c='r',s=80)
    plt.xlabel('x',fontsize=12)
    plt.ylabel('y',fontsize=12)
    plt.xlim(-0,10)
    plt.ylim(-2,2)
    
    kernel = GPy.kern.RBF(input_dim=1, variance=1., lengthscale=1.)
    m = GPy.models.GPRegression(x,function(x),kernel,noise_var=0)
    
    plt.subplots()
    plt.fill_between(x_plot,m.predict_quantiles(x_plot[:,None])[0].flatten(),m.predict_quantiles(x_plot[:,None])[1].flatten(),alpha=0.3)
    plt.plot(x_plot,m.predict(x_plot[:,None])[0])
    plt.plot(x_plot,y_plot,'--',c='C1')
    plt.scatter(x,function(x),marker='x',c='r',s=80)
    plt.xlabel('x',fontsize=12)
    plt.ylabel('y',fontsize=12)
    plt.xlim(-0,10)
    plt.ylim(-2,2)
    x = np.concatenate((x,np.array([(i+1)*0.5])[:,None]),axis=0)
```
